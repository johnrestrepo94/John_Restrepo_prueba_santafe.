**Importar Librerías necesarias**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Importar DataSet**

```python
df = pd.read_csv('GeneraciónCauya2024.csv')
```

```python
df.fillna(0, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Valor_Total'] = df[['Values_Hour01', 'Values_Hour02', 'Values_Hour03', 'Values_Hour04', 'Values_Hour05', 'Values_Hour06', 'Values_Hour07', 'Values_Hour08', 'Values_Hour09', 'Values_Hour10', 'Values_Hour11', 'Values_Hour12', 'Values_Hour13', 'Values_Hour14', 'Values_Hour15', 'Values_Hour16', 'Values_Hour17', 'Values_Hour18', 'Values_Hour19', 'Values_Hour20', 'Values_Hour21', 'Values_Hour22', 'Values_Hour23', 'Values_Hour24', ]].sum(axis=1)
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])


fecha_referencia = pd.Timestamp('2024-12-18')

# Filtrar los últimos 30 días
filtro = (df['Date'] >= fecha_referencia - pd.Timedelta(days=30)) & (df['Date'] <= fecha_referencia)
ultimos_30_dias = df.loc[filtro, ['Date', 'Valor_Total']]

# Mostrar las filas filtradas con fecha y Valor_dia
print(ultimos_30_dias)
```

```python
 #Ordenar por Date de forma ascendente
ultimos_30_dias2 = ultimos_30_dias.sort_values(by='Date', ascending=True)

# Mostrar resultados
print(ultimos_30_dias2)
```

```python
df['Mes_Año'] = df['Date'].dt.to_period('M')

# Calcular la suma de 'Valor_dia' para cada mes
suma_mensual = df.groupby('Mes_Año')['Valor_Total'].sum().reset_index()

# Mostrar el resultado
print(suma_mensual)
```

```python
# Seleccionar las columnas de la 3 a la 27 (índices 2 a 26)
columnas_horas = df.columns[3:26]

# Calcular la suma por columna
suma_por_columna = df[columnas_horas].sum()

# Crear un nuevo DataFrame con los resultados
df_suma = suma_por_columna.reset_index()
df_suma.columns = ['Columna', 'Suma_Total']
```

```python
print(df_suma)
```

```python
valor_mayor = df_suma.loc[df_suma['Suma_Total'].idxmax()]

# Obtener el valor menor
valor_menor = df_suma.loc[df_suma['Suma_Total'].idxmin()]

# Imprimir los resultados
print("Valor mayor:", valor_mayor)
print("Valor menor:", valor_menor)
```

```python
promedio = suma_mensual['Valor_Total'].mean()
desviacion_estandar = suma_mensual['Valor_Total'].std()

# Imprimir los resultados
print(f'Promedio: {promedio}')
print(f'Desviación estándar: {desviacion_estandar}')
```

```python
# Crear un gráfico de barras para 'Valor_Total' por mes
plt.figure(figsize=(10, 6))
sns.barplot(x='Mes_Año', y='Valor_Total', data=suma_mensual, ci=None)
plt.title('Gráfico de Barras - Valor Total por Mes')
plt.xticks(rotation=45)
plt.xlabel('Mes - Año')
plt.ylabel('Valor Total')
plt.tight_layout()
```

```python
# Crear un gráfico de densidad (KDE)
plt.figure(figsize=(10, 6))
sns.kdeplot(suma_mensual['Valor_Total'], shade=True)
plt.title('Gráfico de Densidad de Valor Total')
plt.xlabel('Valor Total')
plt.ylabel('Densidad')
plt.tight_layout()

# Mostrar ambos gráficos
plt.show()
```

```python
from scipy import stats
shapiro_stat, shapiro_p_value = stats.shapiro(suma_mensual['Valor_Total'])
```

```python
print("Prueba de Shapiro-Wilk:")
print(f"Estadístico: {shapiro_stat}, Valor p: {shapiro_p_value}")
```
